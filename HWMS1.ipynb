a) $\theta^* = ((k + 1) \overline{X^k})^{1/k}$ - оценка параметра для равномерного распределения

b) $\theta^* = (\overline{X^k}/k!)^{1/k}$ - оценка параметра для экспоненциального распределения

Нам нужно узнать какой k лучше выбрать, построим графики зависимости среднеквадратического отклонения от k.

```python
import numpy as np
import matplotlib.pyplot as plt
import math


def unif(x_, k):
    return ((k + 1) * x_) ** (1 / k)

def expo(x_, k):
    return (x_ / math.factorial(k)) ** (1 / k)


n = 1000
s = 100
param = 1
```

```python
res_unif = []
for k in range(1, 100):
    deviations = []
    for _ in range(s):
        x_ = sum(np.power(np.random.uniform(0, 1, n), k)) / n
        deviations.append((unif(x_, k) - param) ** 2)
    res_unif.append(sum(deviations) / s)

plt.plot(range(1, 100), res_unif)
plt.title('Uniform')
plt.ylabel('RMSD')
plt.xlabel('k')
plt.show()
```

```python
res_expo = []
for k in range(1, 100):
    deviations = []
    for _ in range(s):
        x_ = sum(np.power(np.random.exponential(1, n), k)) / n
        deviations.append((expo(x_, k) - param) ** 2)
    res_expo.append(sum(deviations) / s)
    
plt.plot(range(1, 100), res_expo)
plt.title('Exponential')
plt.ylabel('RMSD')
plt.xlabel('k')
plt.show()
```

Из полученных графиков сделаем вывод, что для равномерного распределения нужно брать k большое, а у  экспоненциального наоборот поменьше.
